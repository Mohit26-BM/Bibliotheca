# Library Management System

```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv("books_cleaned.csv")

print(df.shape)
print(df.dtypes)
df.head()
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

pd.DataFrame({"missing_count": missing, "missing_%": missing_pct})
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Year as integer
df["published_year"] = pd.to_numeric(df["published_year"], errors="coerce")
df["published_year"] = df["published_year"].fillna(0).astype(int)

# Rating as float
df["average_rating"] = pd.to_numeric(df["average_rating"], errors="coerce")
df["average_rating"] = df["average_rating"].fillna(0.0).round(2)

# Pages and ratings_count as integer
df["num_pages"] = pd.to_numeric(df["num_pages"], errors="coerce").fillna(0).astype(int)
df["ratings_count"] = (
    pd.to_numeric(df["ratings_count"], errors="coerce").fillna(0).astype(int)
)

df.dtypes
```

```python
def fix_isbn(val):
    if pd.isnull(val):
        return ""
    # Convert float like 9.78003e+12 to full string
    try:
        return str(int(float(val)))
    except:
        return str(val).strip()


df["isbn13"] = df["isbn13"].apply(fix_isbn)
df["isbn10"] = df["isbn10"].apply(fix_isbn)

# Preview
df[["isbn13", "isbn10"]].head(10)
```

```python
text_cols = ["title","authors", "categories", "description", "thumbnail"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

# Replace "nan" strings that came from the astype(str) conversion
df[text_cols] = df[text_cols].replace("nan", "")

df.head()
```

```python
df.drop(columns = "subtitle", inplace=True)
df.head()
```

```python
def standardize_delimiters(val, delimiter="/"):
    if not val or val == "":
        return val
    # Replace common delimiters with a single standard one
    val = re.sub(r"\s*[;,&]\s*", delimiter, val)
    return val.strip(delimiter).strip()


df["authors"] = df["authors"].apply(
    lambda x: standardize_delimiters(x, delimiter=" / ")
)
df["categories"] = df["categories"].apply(
    lambda x: standardize_delimiters(x, delimiter=" / ")
)

# Preview unique category values
print(df["categories"].value_counts().head(20))
```

```python
print("Year distribution before:")
print(df["published_year"].describe())

# Keep only sensible years — first book ever printed was ~1450
df_odd_years = df[
    (df["published_year"] != 0)
    & ((df["published_year"] < 1450) | (df["published_year"] > 2025))
]
print(f"\nOutlier years found: {len(df_odd_years)}")
print(df_odd_years[["title", "published_year"]])

df = df[
    (df["published_year"] == 0)
    | ((df["published_year"] >= 1450) & (df["published_year"] <= 2025))
]
print(f"\nRemaining: {len(df)}")
```

```python
print(df["average_rating"].describe())

# Flag contradictory rows: has a rating but 0 ratings_count
contradictory = df[(df["average_rating"] > 0) & (df["ratings_count"] == 0)]
print(f"\nContradictory rating rows: {len(contradictory)}")

# Fix them — zero out the rating
df.loc[(df["average_rating"] > 0) & (df["ratings_count"] == 0), "average_rating"] = 0.0

# Drop ratings outside 1–5 (excluding 0 which means unrated)
df = df[
    (df["average_rating"] == 0)
    | ((df["average_rating"] >= 1) & (df["average_rating"] <= 5))
]

print(f"Remaining: {len(df)}")
```

```python
before = len(df)
df = df.drop_duplicates(subset=["isbn13"], keep="first")
print(f"Dropped {before - len(df)} duplicates")
print(f"Final count: {len(df)}")
```

```python
print("Final shape:", df.shape)
print("\nMissing values:")
print(df.isnull().sum())
print("\nSample:")
df.head()
```

```python
df.to_csv("books_cleaned_final.csv", index=False)
```
